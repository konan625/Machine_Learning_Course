```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.available
plt.style.use('bmh')
```

```python
x_train = np.array([1,2]) 
y_train = np.array([300,500])
```

```python
m = x_train.shape[0]
```

```python
plt.scatter(x_train,y_train,marker='x',c='r')
plt.title("Housing Prices")
plt.ylabel("Prine in 1000s of dollas")
plt.xlabel("Size in 1000sqft")
```

```python
w=100
b=100
```

```python
def compute_model_output(x,w,b):
    m=x.shape[0]
    f_wb = np.zeros(m);
    for i in range(m):
        f_wb[i]=w*x[i]+b
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train,w,b)
plt.plot(x_train,tmp_f_wb,c='r',label='Actual values')
plt.title("Housing Prices")
plt.ylabel("Prine in 1000s of dollas")
plt.xlabel("Size in 1000sqft")
plt.legend()
plt.show()
```
